```python
Services = ['Check balance','Deposit balance','Withdraw balance','Exit']
Balance = 5000

print('Hello! Welcome to our atm banking services: ')

while True:
    
    print()
    print('Please insert your atm card: ')

    print()
    print('Please select an option: ')
    print()
        
    
    for i,j in enumerate(Services,1):
        print(i,j)
    
    option =int(input('\nServices: '))
    
    if option==1:
        print(f'Available balance Rs: {Balance}')
        
    elif option==2:
        num=int(input('How much would you like to deposit: '))
        Balance+=num
        print(f'Your current balance is Rs: {Balance}')
        
    elif option==3:
        num= int(input('How much would you like to withdraw: '))
        Balance-=num
        print(f'Your current balance is Rs: {Balance}')
    elif option==4:
        
        print('Exit')
        
        break
        
print('\nThanks for using our atm banking services')
```
