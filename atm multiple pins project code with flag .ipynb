```python
Customer_Info = { 1999 :['User1', 1000], 1111 :['User2', 5000],
                 3333 :['User3', 10000]}

Services = ['Check balance','Deposit balance','Withdraw balance','Exit']

Enter_pin = True

while Enter_pin:
    print()
    print('\nPlease insert your atm card: ')

    
    Enter_pin =int(input('Enter your pin: '))
    
    if Enter_pin in Customer_Info.keys():
        
        print(f'Pin correct , Hello! {Customer_Info[Enter_pin][0]} Welcome to our atm banking services! ')
        
        while True:
            
            print()
            print('Please select an option: ')
            print()
            
            

        
    
            for i,j in enumerate(Services,1):
              print(i,j)
    
            option =int(input('\nServices: '))
    
            if option==1:
                print(f'Available balance Rs: {Customer_Info[Enter_pin][1]}')
        
            elif option==2:
                num=int(input('How much would you like to deposit: '))
                Customer_Info[Enter_pin][1]+=num
                print(f'Your current balance is Rs: {Customer_Info[Enter_pin][1]}')
        
            elif option==3:
                num= int(input('How much would you like to withdraw: '))
                Customer_Info[Enter_pin][1]-=num
                print(f'Your current balance is Rs: {Customer_Info[Enter_pin][1]}')
            elif option==4:
        
                print('Exit')
                print()
                print("Thanks for using our atm banking services!")
                
                Enter_pin = False
        
                break

                
    else:
        print('Incorrect Pin, Please try again: ')
    
```
