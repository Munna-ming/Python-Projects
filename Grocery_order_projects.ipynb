```python
home_essentials = ['chips','cake','bucket','plate','cup','spoon','soap','bowl','mats','bottle']
item_prices = {'chips':10,'cake':20,'bucket':180,'plate':200,'cup':50,'spoon':30,'soap':20,'bowl':40,'mats':100,'bottle':250}

total_amount = []
final_order = {}
customer_address ={} 

print('Hi! Welcome to our text based home_essentials ordering system.')

order_item = True

while order_item:
    print('Please select an item : ')
    print()
    
    for item in home_essentials:
        print()
        print(home_essentials)
        print()
        break
        
    while True:
        
        item = input('\nWhich item would you like to order: ')
        

        if item in home_essentials:
            print(f'You have selected product: {item}')
            final_order.setdefault(item,[])
            break
             
        if item not in home_essentials:
            print(f'Sorry we do not have {item} product at the moment')
           
    while True:
        
        extra_items=input('Would you like to add extra_item ? Yes or No: ')
    
        if extra_items=='Yes' or extra_items=='yes':
            
            product = input('Which item would you like to add ? : ')
 
            if product in home_essentials:
               print(f'You have selected product {product} .')
 
               final_order[item].append(product)
 
            else:
                print(f'Sorry we do not have product {item}  at the moment.')
 
        elif extra_items=='No' or extra_items=='no':
        
           break
 
        check_order = True    
  
        while check_order:
            print()
            
            order_correct = input('Is this correct? Yes or No: ')
            if order_correct.lower()=='yes':
                check_order=False
                order_item=False
                
            if order_correct.lower()=='no':
                print(final_order)
                
            add_remove = input('Would you like to add or remove? Yes or No: ')
                
            if add_remove.lower()=='remove':
                
                remove=input('Which home_essentials would you like to remove? ')
                del final_order[remove]
                print(final_order)
            if add_remove.lower()=='add':
                check_order=False
            
print()    
        
print(final_order)
item_order=[]
product_order=[]
for i in final_order.keys():
    item_order.append(i) 
print()
for k in final_order.keys():
    for j in final_order.get(k):

        product_order.append(j)
for sw in item_order:
    total_amount.append(item_prices.get(sw))
    for fil in product_order:
        total_amount.append(item_prices.get(fil))
print(f'\nYour total order price is Rs. {sum(total_amount)}')
print("Please provide us with your name, address and phone number.")
customer_address['name']=input('Your name please: ')
customer_address['street_name']=input('Street name and house number: ')
customer_address['postalcode']=input('Postal code and city name: ')
customer_address['phonenumber']=input('Your number: ')
print()
print(f"Thank you for your order {customer_address['name']}.")
print()
print(f'We will deiver your order to this address.')
print()
print(customer_address['street_name'])
print(customer_address['postalcode'])
print()
print(f"We will contact you on {customer_address['phonenumber']} if anything comes")
    
            
    
```
